```python

from llama_index.embeddings.huggingface import HuggingFaceEmbedding
# from llama_index.llms.openai import OpenAI
from llama_index.core import Settings
from llama_index.postprocessor.flag_embedding_reranker import (
    FlagEmbeddingReranker,
)
from llama_index.core.node_parser import SemanticSplitterNodeParser
from llama_index.core.ingestion import IngestionPipeline
import chromadb
from llama_index.vector_stores.chroma import ChromaVectorStore
from llama_index.core import VectorStoreIndex
import torch
from dotenv import load_dotenv
```

```python
torch.cuda.is_available()  # Check if CUDA is available
load_dotenv()
```

```python
#Settings.embed_model = HuggingFaceEmbedding(
embed_model = HuggingFaceEmbedding(
    model_name="BAAI/bge-m3",
    device="cuda"
)
rerank = FlagEmbeddingReranker(model="BAAI/bge-reranker-large", top_n=20)
print("Embedding model and reranker initialized.")
```

```python
splitter = SemanticSplitterNodeParser(
    buffer_size=1, breakpoint_percentile_threshold=95, embed_model=embed_model
)
```

```python
fyp1_client = chromadb.PersistentClient(path="./prototype_fyp1")  # Persistent storage
collection = fyp1_client.get_or_create_collection(name="fyp1")
vector_store = ChromaVectorStore(chroma_collection=collection, embed_model=embed_model)
```

```python
pipeline = IngestionPipeline(
    transformations=[
        splitter,
        embed_model,
    ],
    vector_store=vector_store,
)
```

```python
pipeline.run(documents=doc)
print("Ingestion pipeline completed.")
```

```python
index = VectorStoreIndex.from_vector_store(vector_store)
```
